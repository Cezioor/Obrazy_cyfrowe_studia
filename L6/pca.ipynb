```python
from skimage import io, color, morphology, filters, data, exposure, segmentation, measure, feature
from matplotlib import pyplot as plt
import numpy as np
from sklearn.decomposition import PCA


img = io.imread('sample.jpg')
img = color.rgb2gray(img)


# Parametry
num_components = 25

pca = PCA(n_components=num_components)
img_transformed = pca.fit_transform(img)        # (H, num_components)
img_reconstructed = pca.inverse_transform(img_transformed)  # (H, W)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.title("Oryginalny obraz")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title(f"PCA rekonstrukcja ({num_components} skadowych)")
plt.imshow(img_reconstructed, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()

```
