```python
### import numpy as np
import numpy as np
from numpy import pi
import sounddevice as sd
import matplotlib.pyplot as plt
fs = 10000 # Hz

dt=1/fs

t=np.arange(0,1,dt)
t2=np.arange(0,1/2.,dt)
t4=np.arange(0,1/4.,dt)
t8=np.arange(0,1/8.,dt)
#f=500
fc=261.63;
fd=293.66;
fe=329.631;
ff=349.23;
fg=392.00;

#y=np.sin(2*np.pi*f*t)
xg=np.sin(2*pi*fg*t4);
xe=np.sin(2*pi*fe*t4);
xf=np.sin(2*pi*ff*t4);
xd=np.sin(2*pi*fd*t4);
xc8=np.sin(2*pi*fc*t8);
xe8=np.sin(2*pi*fe*t8);
xg2=np.sin(2*pi*fg*t2);
xc2=np.sin(2*pi*fc*t2);
#plt.plot(t[0:5000],xg[0:5000],'x-')
x=np.concatenate((xg,xe, xe, xf, xd, xd, xc8, xe8, xg2 ,xg ,xe, xe ,xf, xd, xd ,xc8 ,xe8 ,xc2));
#x=np.concatenate((xd, 2*xd ));


sd.play(x,fs)
xtmp=xg+xe+xf#+xd
#sd.play(x,fs)
#Fw=np.fft.fft(xtmp)
Fw=np.fft.fft(x)
Fws=np.fft.fftshift(Fw)
f=np.linspace(-0.5,0.5,Fw.shape[0])*fs
plt.plot(f,np.abs(Fws))
plt.xlim((200,500))
plt.show()
```
